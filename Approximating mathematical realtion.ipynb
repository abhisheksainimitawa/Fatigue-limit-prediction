### For Model 1

#### 1. Exact solution

```python
import tensorflow as tf
import numpy as np

# Load your trained model
model = tf.keras.models.load_model('my_model_1.h5')

# Extract weights and biases from the trained model
weights = model.get_weights()

# Define the input variables
x1, x2, x3 = "x1", "x2", "x3"

# Define ReLU function
def relu(x):
    return np.maximum(0, x)

# Define the equations for each layer
equations = []

# Input layer to first hidden layer
for j in range(8):
    equation = f"h1_{j+1} = relu({'+'.join([f'({weights[0][i, j]}*{x})' for i, x in enumerate([x1, x2, x3])])} + {weights[1][j]})"
    equations.append(equation)

# First hidden layer to second hidden layer
for j in range(7):
    equation = f"h2_{j+1} = relu({'+'.join([f'({weights[2][i, j]}*h1_{i+1})' for i in range(8)])} + {weights[3][j]})"
    equations.append(equation)

# Second hidden layer to third hidden layer
for j in range(4):
    equation = f"h3_{j+1} = relu({'+'.join([f'({weights[4][i, j]}*h2_{i+1})' for i in range(7)])} + {weights[5][j]})"
    equations.append(equation)

# Third hidden layer to output layer
y_equation = f"y = {'+'.join([f'({weights[6][j]}*h3_{j+1})' for j in range(4)])} + {weights[7][0]}"
equations.append(y_equation)

# Combine all equations
final_equation = '\n'.join(equations)

print("Equation:")
print(final_equation)


```
